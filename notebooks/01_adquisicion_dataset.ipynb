```python
# 01_adquisicion_dataset.ipynb
# Adquisicion del dataset: Top Spotify Songs in 73 Countries
# Descarga, carga y verificacion inicial del dataset

import os
import pandas as pd
import kagglehub

# Configuracion de rutas
RAW_DATA_PATH = "./../data/raw/universal_top_spotify_songs.csv"
PROCESSED_DATA_PATH = "./../data/processed/"

def cargar_o_descargar_dataset(ruta_local, dataset_handle="asaniczka/top-spotify-songs-in-73-countries-daily-updated"):
    """
    Carga el dataset desde archivo local si existe.
    Si no existe, lo descarga desde Kaggle y lo guarda en la ruta especificada.
    """
    # Verificar si el archivo ya existe localmente
    if os.path.exists(ruta_local):
        print("Archivo encontrado localmente:", ruta_local)
        df = pd.read_csv(ruta_local)
        print("Dimensiones del dataset:", df.shape)
        return df

    # Si no existe, descargar desde Kaggle
    print("Archivo no encontrado en:", ruta_local)
    print("Descargando desde Kaggle:", dataset_handle)

    try:
        # Descargar dataset a la cache de Kaggle
        download_path = kagglehub.dataset_download(dataset_handle)
        print("Dataset descargado en:", download_path)

        # Buscar el archivo CSV en la carpeta descargada
        archivos_csv = [f for f in os.listdir(download_path) if f.endswith('.csv')]

        if not archivos_csv:
            raise FileNotFoundError("No se encontró ningun archivo CSV en la descarga")

        archivo_csv = archivos_csv[0]
        ruta_completa = os.path.join(download_path, archivo_csv)

        # Cargar el CSV
        df = pd.read_csv(ruta_completa)
        print("Dataset cargado correctamente. Dimensiones:", df.shape)

        # Crear directorio raw si no existe y guardar copia local
        os.makedirs(os.path.dirname(ruta_local), exist_ok=True)
        df.to_csv(ruta_local, index=False)
        print("Copia guardada localmente en:", ruta_local)

        return df

    except Exception as e:
        print("Error al descargar o cargar el dataset:", e)
        raise

# Crear directorio processed si no existe
os.makedirs(PROCESSED_DATA_PATH, exist_ok=True)

# Ejecutar la carga
df = cargar_o_descargar_dataset(RAW_DATA_PATH)

# Verificacion rapida
print("\nPrimeras filas:")
print(df.head())
print("\nColumnas disponibles:")
print(df.columns.tolist())
```
