```python
# ============================================
# 04_muestreo_escalado.ipynb
# Preparacion de datos para clustering
# ============================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib

# ============================================
# CONFIGURACION
# ============================================
RUTA_DATA = "./../data/processed/"
TAMANO_MUESTRA = 100000
RANDOM_STATE = 42

# ============================================
# 1. CARGAR DATOS LIMPIOS
# ============================================
print("Cargando datos limpios...")
print("Usando spotify_features_clean.csv (solo las 11 columnas)")

df = pd.read_csv(RUTA_DATA + "spotify_features_clean.csv")
print(f"Dataset cargado: {len(df):,} canciones")
print(f"Columnas disponibles: {df.shape[1]}")

# ============================================
# 2. MUESTREO
# ============================================
print("\n" + "="*50)
print("PASO 1: MUESTREO")
print("="*50)

df_muestra = df.sample(n=min(TAMANO_MUESTRA, len(df)), random_state=RANDOM_STATE)
print(f"Muestra tomada: {len(df_muestra):,} canciones")
print(f"Porcentaje del total: {len(df_muestra)/len(df)*100:.1f}%")

# Guardar muestra sin escalar (opcional, por si se necesita despues)
df_muestra.to_csv(RUTA_DATA + "spotify_muestra_100k_raw.csv", index=False)
print("Archivo guardado: spotify_muestra_100k_raw.csv")

# ============================================
# 3. ESCALADO
# ============================================
print("\n" + "="*50)
print("PASO 2: ESCALADO CON STANDARDSCALER")
print("="*50)

features = df.columns.tolist()
print(f"Variables a escalar: {len(features)}")
for i, f in enumerate(features, 1):
    print(f"  {i}. {f}")

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_muestra[features])

print(f"\nDatos escalados correctamente")
print(f"Media resultante: {X_scaled.mean(axis=0).round(2)}")
print(f"Desviacion resultante: {X_scaled.std(axis=0).round(2)}")

# ============================================
# 4. GUARDAR ARCHIVOS
# ============================================
print("\n" + "="*50)
print("PASO 3: GUARDAR ARCHIVOS")
print("="*50)

df_scaled = pd.DataFrame(X_scaled, columns=features)
df_scaled.to_csv(RUTA_DATA + "spotify_muestra_100k_scaled.csv", index=False)
print("Guardado: spotify_muestra_100k_scaled.csv")

joblib.dump(scaler, RUTA_DATA + "scaler.pkl")
print("Guardado: scaler.pkl")

# ============================================
# 5. VERIFICACION FINAL
# ============================================
print("\n" + "="*50)
print("RESUMEN FINAL")
print("="*50)

print(f"""
Archivos generados en {RUTA_DATA}:

1. spotify_muestra_100k_raw.csv
\t- 100,000 canciones
\t- Datos originales sin escalar
\t- Para referencia o EDA adicional

2. spotify_muestra_100k_scaled.csv
\t- 100,000 canciones
\t- Datos escalados (media=0, std=1)
\t- ESTE ES EL ARCHIVO PARA CLUSTERING

3. scaler.pkl
\t- Objeto StandardScaler entrenado
\t- Para escalar nuevos datos en el futuro
""")

print("Datos listos para clustering")
```
